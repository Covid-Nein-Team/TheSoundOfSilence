```python
import pandas as pd
import datetime
from sklearn.preprocessing import RobustScaler
from miditime.miditime import MIDITime
```

```python
covid_file="covid_19_cases.csv"
covid_data=pd.read_csv(covid_file,index_col='date', parse_dates=True)
covid_data.iloc[0]=0.0
covid_data.interpolate(method='time', inplace=True)    
covid_data
```

```python
def mag_to_pitch_tuned(magnitude, min_value, max_value):
    # Where does this data point sit in the domain of your data? (I.E. the min magnitude is 3, the max in 5.6). In this case the optional 'True' means the scale is reversed, so the highest value will return the lowest percentage.
    scale_pct = mymidi.linear_scale_pct(min_value, max_value, magnitude)

    # Another option: Linear scale, reverse order
    # scale_pct = mymidi.linear_scale_pct(3, 5.7, magnitude, True)

    # Another option: Logarithmic scale, reverse order
    #scale_pct = mymidi.log_scale_pct(3, 5.7, magnitude, True)

    # Pick a range of notes. This allows you to play in a key.
    #Phrygian
    scale = ['E', 'F', 'G', 'A', 'B', 'C', 'D' ,'E']
    #c-major
    #scale = ['C', 'D', 'E', 'F', 'G', 'A', 'B']

    #Find the note that matches your data point
    note = mymidi.scale_to_note(scale_pct, scale)

    #Translate that note to a MIDI pitch
    midi_pitch = mymidi.note_to_midi_pitch(note)

    return midi_pitch
```

```python
countries = covid_data.columns.tolist()
for country in countries:
    print(country)
    min_value = min(covid_data[country])
    max_value = max(covid_data[country])
    data_one_country = covid_data[country].tolist()
    my_data_epoched = []
    mymidi = MIDITime(120, country + '.mid', 20, 5, 1)

    for idx,data in enumerate(data_one_country):
        my_data_epoched.append({'days_since_epoch': idx,'magnitude': data})
 
    my_data_timed = [{'beat': mymidi.beat(d['days_since_epoch']), 'magnitude': d['magnitude']} for d in my_data_epoched]
    start_time = my_data_timed[0]['beat']
    
    note_list = []

    for d in my_data_timed:
        note_list.append([
        d['beat'] - start_time,
        mag_to_pitch_tuned(d['magnitude'], min_value, max_value),
        100,  # velocity
        1  # duration, in beats
    ])
    # Add a track with those notes
    mymidi.add_track(note_list)

    # Output the .mid file
    mymidi.save_midi()
```
